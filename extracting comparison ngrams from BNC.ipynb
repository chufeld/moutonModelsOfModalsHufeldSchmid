## unigram comparisons: take and get

```python
unigramComparisonList = []
with concurrent.futures.ProcessPoolExecutor() as executor:
    for i in df.itertuples():
        if i.token == "take" or i.token == "get":
            unigramComparisonList.append(i)
```
